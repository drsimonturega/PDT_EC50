# EC50 calculation for PDT treatment of cancer cells with a PS


## uploading data table

```python
#upload *.csv
import os
import pandas as pd
import numpy as np
import statistics

class ec50_anal():
    
    def __init__(exp_name):
        #self.exp_name = exp_name
        return
        
    def load_and_vis(exp_name):
        #upload *.csv
        try:
            df = pd.read_csv(f'{exp_name}.csv', encoding="utf-8")
        except PermissionError:
            print('The file we are trying to open may be open in another \
                program, please close and try again!')
        
        # taking averages
        mean_plot = pd.DataFrame()
        #mean_plot = []
        col_names = list(df.columns.values)
        print(col_names)
        num_con = 5
        num_rep = 3
        num_plate = 3
        #cur_col = 3
        for cur_col in range(3,7):
            mean_plot.insert(cur_col-3, col_names[cur_col], 0,)
            for n in range(0, (num_con + 1) * num_plate):
            
                mean_plot.at[n, col_names[cur_col]] = statistics.mean(df.iloc[(n*num_rep):(n*num_rep)+(num_rep-1), cur_col])
        for cur_col in range(0,3):
            mean_plot.insert(cur_col, col_names[cur_col], df[col_names[cur_col]])
        print(df.head(10))
        return mean_plot.head(10)

first_try = ec50_anal()
ec50_anal.load_and_vis('EC5_H28_22042024')
```
